```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
#https://pytorch.org/hub/ultralytics_yolov5/ aca puedes ver cuales modelos ocupar s(small)
model=torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
#ruta a una imagen
img='http://implansaltillo.mx/img_articulos_revista/articulo_40.jpg'
#pasar imagen por la red
results=model(img)
#mostar resultados en texto 
results.print()
```

```python
#visualizacion grafica de resultados
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
#usar red para  desde webcam
cap = cv2.VideoCapture(1)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
